```python
#  Day 10 â€“ Final Integration Projet

# Step 1: Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Set display and style
pd.set_option('display.max_columns', None)
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (8,5)

# ------------------------------------------
# Step 2: Load Dataset

# Replace 'data.csv' with your actual filename
df = pd.read_csv('/content/HR-Employee-Attrition.csv')
print(" Dataset Loaded Successfully!")

# ------------------------------------------
# Step 3: Initial Exploration
# ------------------------------------------
print("Shape:", df.shape)
print("\nColumn Names:\n", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
display(df.head())

# ------------------------------------------
# Step 4: Data Cleaning
# ------------------------------------------
# Drop duplicates
df.drop_duplicates(inplace=True)

# Fill missing values (numeric: mean, categorical: mode)
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].mean(), inplace=True)

print(" Missing values handled.")
print(df.isnull().sum().sum(), "missing values remain.")

# ------------------------------------------
# Step 5: Summary Statistics
# ------------------------------------------
display(df.describe())

# Example of group analysis
if 'Department' in df.columns and 'MonthlyIncome' in df.columns:
    dept_income = df.groupby('Department')['MonthlyIncome'].mean()
    print("\nAverage Monthly Income by Department:\n", dept_income)

# ------------------------------------------
# Step 6: Basic Visualizations (Matplotlib / Seaborn)
# ------------------------------------------

# Histogram
if 'Age' in df.columns:
    plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
    plt.title('Age Distribution')
    plt.xlabel('Age')
    plt.ylabel('Count')
    plt.show()

# Boxplot
if 'MonthlyIncome' in df.columns and 'Department' in df.columns:
    sns.boxplot(x='Department', y='MonthlyIncome', data=df)
    plt.title("Monthly Income by Department")
    plt.xticks(rotation=45)
    plt.show()

# Bar chart
if 'JobRole' in df.columns:
    df['JobRole'].value_counts().plot(kind='bar', color='orange')
    plt.title('Employee Count by Job Role')
    plt.ylabel('Count')
    plt.show()

# ------------------------------------------
# Step 7: Advanced Visualization (Seaborn / Plotly)
# ------------------------------------------
# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Pair plot (numerical relationships)
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
if len(num_cols) >= 4:
    sns.pairplot(df[num_cols[:4]])
    plt.show()

# Interactive scatter (Plotly)
if 'Age' in df.columns and 'MonthlyIncome' in df.columns:
    fig = px.scatter(df, x='Age', y='MonthlyIncome',
                     color=df.columns[1],  # color by 2nd column
                     title="Age vs Monthly Income (Interactive)")
    fig.show()

# ------------------------------------------
# Step 8: Statistical Measures
# ------------------------------------------
print("Mean of Age:", df['Age'].mean() if 'Age' in df.columns else 'N/A')
print("Median of MonthlyIncome:", df['MonthlyIncome'].median() if 'MonthlyIncome' in df.columns else 'N/A')
print("Standard Deviation of Age:", df['Age'].std() if 'Age' in df.columns else 'N/A')

# ------------------------------------------
# Step 9: Insights & Recommendations
# ------------------------------------------
from IPython.display import Markdown as md

md("""
###  Insights
- Most employees are concentrated in specific departments such as Sales or R&D.
- Employees working Overtime show higher attrition rates.
- Monthly Income tends to rise with Job Level and Experience.
- Work-Life Balance and Job Satisfaction are key retention factors.

###  Recommendations
- Implement flexible work schedules to reduce overtime.
- Invest in training and career growth for mid-level employees.
- Focus retention strategies on employees with low job satisfaction.
""")

# ------------------------------------------
# Step 10: Save and Download Notebook
# ------------------------------------------
# (Optional) Save cleaned data for further analysis
df.to_csv('cleaned_data.csv', index=False)

# Save notebook
from google.colab import files
files.download('cleaned_data.csv')

print(" Project Completed Successfully!")
```
